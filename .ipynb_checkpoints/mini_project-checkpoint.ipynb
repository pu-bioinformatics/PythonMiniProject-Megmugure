```python
def menu(option = "None", choice = "  None  "):
    
    option = input(":")
    """
    This function prints out the display menu.
    """
    print("********************************************************************************")
    print("* PDB FILE ANALYZER                                                            *")
    print("********************************************************************************")
    print("* Select an option from below:                                                 *")
    print("*                                                                              *")
    print("*      1)Open a PDB file                        (O)                            *")
    print("*      2)Information                            (I)                            *")
    print("*      3)Show histogram of amino acids          (H)                            *")
    print("*      4)Display secondary structures           (S)                            *")          
    print("*      5)Expoert PDB File                       (X)                            *")
    print("*      6)Exit                                   (Q)                            *")
    print("*                                                                              *")
    print("*                                                   Current PDB: %s      *" % choice)
    print("********************************************************************************")
    print(": %s " % option)
    
    
menu()
```

```python
def choice():
    choice = input(":")
    if choice.upper() in ("O", "I", "H", "S", "X", "Q"):
        if choice.upper() == "O":
            option = input("Enter a Valid PATH for a PDB file: ")
            print("The File " + option + " has been successfully loaded")
            menu(choice, option)   
        elif choice.upper() == "I":
            print("PDB File: 3AYU.pdb")
            information()
            menu()
        elif choice.upper() == "H":
            hist_option()
            menu()
        elif choice.upper() == "S":
            print("Secondary Structure of the PDB id 3AYU:")
            secondary()
            menu()
        elif choice.upper() == "X":
            print("Export PDB File")
        else:
            exit()
        
            
choice()
```

```python
dict4 = {'CYS': 'C','ASP': 'D','SER': 'S','GLN': 'Q','LYS': 'K','ILE': 'I','PRO': 'P','THR': 'T','PHE': 'F','ASN': 'N','GLY': 'G','HIS': 'H','LEU': 'L','ARG': 'R','TRP': 'W','ALA': 'A','VAL': 'V','GLU': 'E','TYR': 'Y','MET': 'M'}
```

```python

def information():
    """
    This function displays the title of the pdb file, the name of
    the file as well as a summary of the file.
    """
    with open("3AYU.pdb", 'r') as my_file:
        lines = my_file.readlines()
        
        title_lst = []
        chains = []
        for line in lines:
            split_lines = line.split()
            if line.startswith("TITLE"):
                split_lines = split_lines[1:]
                for words in split_lines:
                    title_lst.append(words)
                    title = " ".join(title_lst)
                    title = "Title: " + title

            if line.startswith("SEQRES"):
                if split_lines[2] not in chains:
                    chains.append(split_lines[2])
                    extracted_chains = " and ".join(chains)
                    extracted_chains = "CHAINS: " + extracted_chains

        print(title)
        print(extracted_chains)

        for chain in chains:
            print("-Chain " + chain)

            amino_list = []
            num_string = ""
            helix = 0
            sheet = 0
            for line in lines:
                split_lines = line.split()

                if line.startswith("SEQRES"):
                    if chain == split_lines[2]:

                        num_string = "  Number of amino acids:   " + str(split_lines[3])
                        for amino in split_lines[4:]:
                            amino_list.append(dict4[amino])
                    

                if line.startswith("HELIX"):
                    if chain == split_lines[4]:
                        helix = helix + 1

                if line.startswith("SHEET"):
                    if chain == split_lines[5]:
                        sheet = sheet + 1

            print(num_string)
            print("  Number of helix:          " + str(helix))
            print("  Number of sheet:          " + str(sheet))
            

            amino_list = "".join(amino_list)
            sequence_lst = []
            for i in range(0, len(amino_list), 50):
                seq_string = "            "+amino_list[i:i+50]
                sequence_lst.append(seq_string)
            line1 = list(sequence_lst[0])
            line1[2:12] = "Sequence: "
            line1 = "".join(line1)
            sequence_lst[0] = line1
            for line in sequence_lst:
                print(line)
            
            
            
                
            
            
                   
                


information()    
```

```python
def hist(f = "None"):
    input(":")
    input("Choose an option to order by: ")
    print("   number of amino acids - ascending   (an)")
    print("   number of amino acids - descending  (dn)")
    print("   alphabetically - ascending          (aa)")
    print("   alphabetically - descending         (da)")
    print("order by: %s" %f)
        
hist()
```

```python
dict1 = {'ASP': 'D','SER': 'S','GLN': 'Q','LYS': 'K','ILE': 'I','PRO': 'P','THR': 'T','PHE': 'F','ASN': 'N','GLY': 'G','HIS': 'H','LEU': 'L','ARG': 'R','TRP': 'W','ALA': 'A','VAL': 'V','GLU': 'E','TYR': 'Y','MET': 'M'}
```

```python
def hist_option():
    option = input("Select an option to order by: ")
    hist()
    option = option.lower()
    if option not in ("an", "dn", "aa", "da"):
        print("Invalid option")
        hist_option()
        
    else:
        if option == "an":
            ascend_nums()
        if option == "dn":
            descend_nums()
        if option == "da":
            descend_alphabetic()
        if option == "aa":
            ascend_alphabetic()
        
            
            
            
hist_option()
```

```python
def ascend_alphabetic():
    """
    This function enables the ordering of the amino acids in the alphabetic order,
    in the ascending order, from a to z.
    """
    with open ("3AYU.pdb", "r") as file:
        lines = file.readlines()
        amino_acids = []
        
        for line in lines:
            split_lines = line.split()
            
            if line.startswith("SEQRES"):
                for amino in split_lines[4:]:
                    amino_acids.append(amino)
                    
                    dict2 = dict((i, 0) for i in dict1.keys())
                    for amino in amino_acids:
                        dict2[amino] = dict2[amino] + 1
                        sorted_amino = sorted(dict2)
                        
        for i in sorted_amino:
            print("%s (%3d): %s" %(i, dict2[i], '*'*dict2[i]))        
                    
ascend_alphabetic()
```

```python
def descend_alphabetic():
    """
    This function enables the ordering of the amino acids in the alphabetic order,
    in the descending order, from z to a.
    """
    with open ("3AYU.pdb", "r") as file:
        lines = file.readlines()
        amino_acids = []
        
        for line in lines:
            split_lines = line.split()
            
            if line.startswith("SEQRES"):
                for amino in split_lines[4:]:
                    amino_acids.append(amino)
                    
                    dict2 = dict((i, 0) for i in dict1.keys())
                    for amino in amino_acids:
                        dict2[amino] = dict2[amino] + 1
                        sorted_amino = sorted(dict2, reverse = True)
                        

        
        for i in sorted_amino:
            print("%s (%3d): %s" %(i, dict2[i], '*'*dict2[i]))
            
                    
descend_alphabetic()
```

```python
def ascend_nums():
    """
    This function enables the ordering of the amino acids according to their 
    numbers, in the ascending order, from the lowest to the highest number.
    """
    with open ("3AYU.pdb", "r") as file:
        lines = file.readlines()
        amino_acids = []
        
        for line in lines:
            split_lines = line.split()
            
            if line.startswith("SEQRES"):
                for amino in split_lines[4:]:
                    amino_acids.append(amino)
                    
                    dict2 = dict((i, 0) for i in dict1.keys())
                    for amino in amino_acids:
                        dict2[amino] = dict2[amino] + 1
                        sorted_amino = dict(sorted(dict2.items(), key = lambda t: t[1]))
                        
        for i in sorted_amino:
            print("%s (%3d): %s" %(i, dict2[i], '*'*dict2[i]))
            
                    
ascend_nums()
```

```python
def descend_nums():
    """
    This function enables the ordering of the amino acids according to their 
    numbers, in the descending order, from the highest to the lowest number.
    """
    with open ("3AYU.pdb", "r") as file:
        lines = file.readlines()
        amino_acids = []
        
        for line in lines:
            split_lines = line.split()
            
            if line.startswith("SEQRES"):
                for amino in split_lines[4:]:
                    amino_acids.append(amino)
                    
                    dict2 = dict((i, 0) for i in dict1.keys())
                    for amino in amino_acids:
                        dict2[amino] = dict2[amino] + 1
                        sorted_amino = dict(sorted(dict2.items(), key = lambda t: t[1], reverse = True))
                       
        for i in sorted_amino:
            print("%s (%3d): %s" %(i, dict2[i], '*'*dict2[i]))
            
                    
descend_nums()
```

```python
dict4 = {'CYS': 'C','ASP': 'D','SER': 'S','GLN': 'Q','LYS': 'K','ILE': 'I','PRO': 'P','THR': 'T','PHE': 'F','ASN': 'N','GLY': 'G','HIS': 'H','LEU': 'L','ARG': 'R','TRP': 'W','ALA': 'A','VAL': 'V','GLU': 'E','TYR': 'Y','MET': 'M'}
```

```python
def secondary():
    """
    This function displays the secondary structure of the pdb file.
    """
    with open('3AYU.pdb', 'r') as file:
        lines = file.readlines()

        chains = []
        amino_lst = []
        for line in lines:
            split_lines = line.split()
            if line.startswith("SEQRES"):
                if split_lines[2] not in chains:
                    chains.append(split_lines[2])

        for chain in chains:
            print("Chain " + chain + ":" + "\n" + "(1)")

            amino_lst = []
            empty_amino_lst = []
            empty_dashes_lst = []
            for line in lines:
                split_lines = line.split()

                if line.startswith("SEQRES"):
                    if chain == split_lines[2]:
                        for amino in split_lines[4:]:
                            amino_lst.append(dict4[amino])
                            empty_amino_lst.append(" ")
                            empty_dashes_lst.append("-")
                            
                if line.startswith("HELIX"):
                    if chain == split_lines[4]:
                        
                        helix_name = split_lines[2]
                        empty_amino_lst[int(split_lines[5]) - 1 : int(split_lines[5]) - 1 + len(helix_name)] = helix_name
                        
                        x = int(split_lines[8]) - int(split_lines[5]) + 1
                        y = x * "/"
                        empty_dashes_lst[int(split_lines[5]) - 1 : int(split_lines[8])] = y
                        
                        
                                 
                        
                if line.startswith("SHEET"):
                    if chain == split_lines[5]:
                        
                        sheet_name = split_lines[1] + split_lines[2]
                        empty_amino_lst[int(split_lines[6]) - 1 : int(split_lines[6]) - 1 + len(sheet_name)] = sheet_name
                        
                        x = int(split_lines[9]) - int(split_lines[6]) + 1
                        n = x * "|"
                        empty_dashes_lst[int(split_lines[6]) - 1 : int(split_lines[9])] = n
                        
            
            #l = str(len(empty_amino))
            #print(l)
            amino_lst = "".join(amino_lst)
            empty_dashes_lst = "".join(empty_dashes_lst)
            
            empty_amino_lst = "".join(empty_amino_lst)
            
            
            
    
            for i in range(0, len(amino_lst), 80):
                print(amino_lst[i:i+80], "\n" + empty_dashes_lst[i:i+80] + "\n" + empty_amino_lst[i:i+80])
            
            
secondary()  
```
